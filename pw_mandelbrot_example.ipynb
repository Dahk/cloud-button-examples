```python
from pw_mandelbrot import mandelbrot_image

width = height = 512
maxiter = 60
concurrency = 1024  # number of functions executed in parallel
```

```python
%matplotlib notebook

# Initial rectangle position
xmin=-2.0
dx=2.5
ymin=-1.25
dy=2.5

subplots = mandelbrot_image(xmin, xmin+dx, ymin, ymin+dy, width, height, maxiter, concurrency)

# 1st zoom
delta = dx * 0.5
xmin = xmin
ymin = ymin + dy * 0.5 - delta * 0.5
dx = dy = delta
mandelbrot_image(xmin, xmin+dx, ymin, ymin+dy, width, height, maxiter, concurrency, subplots)

# 2nd zoom
delta = dx * 0.3666
xmin = xmin
ymin = ymin + dy * 0.5 - delta * 0.5
dx = dy = delta
mandelbrot_image(xmin, xmin+dx, ymin, ymin+dy, width, height, maxiter, concurrency, subplots)

# 3rd zoom
delta = dx * 0.3333
xmin = xmin + dx * 0.3333
ymin = ymin + dy * 0.5 - delta * 0.5
dx = dy = delta
mandelbrot_image(xmin, xmin+dx, ymin, ymin+dy, width, height, maxiter, concurrency, subplots)

# 4th zoom
delta = dx * 0.3333
xmin = xmin + dx * 0.3333
ymin = ymin + dy * 0.5 - delta * 0.5
dx = dy = delta
mandelbrot_image(xmin, xmin+dx, ymin, ymin+dy, width, height, maxiter, concurrency, subplots)
```
